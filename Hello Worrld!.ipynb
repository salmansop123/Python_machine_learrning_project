```python
# Model making ad training
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #import model to train
from sklearn.model_selection import train_test_split #import function to split our data to train and test
from sklearn.metrics import accuracy_score

music_data = pd.read_csv("music.csv")

X = music_data.drop(columns=["genre"]) #We describe input set using capital X 
y = music_data["genre"] #We describe output set using small y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) #In this we tell how much to split data to train our model


model = DecisionTreeClassifier()
model.fit(X_train ,y_train) # giving data to train our model
predictions = model.predict(X_test)#store the perdiction of the model

score = accuracy_score(y_test, predictions) #Calculate the accuracy of the model
score
```
